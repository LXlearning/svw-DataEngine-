# 汽车质量数据统计
Tips
*  ../ 表示当前文件所在的目录的上一级目录 
* ./ 表示当前文件所在的目录(可以省略) 
* / 表示当前站点的根目录(域名映射的硬盘目录)

```python
data = pd.read_csv(path+'car_complain.csv',sep=',')
```

```python
data.head()
```

```python
data['problem'].str.split(',',expand=True).head()##.loc[:,8].value_counts()
```

```python
data['brand'].value_counts().index[0]
```

```python
data['car_model'].value_counts().index[0]
```

```python
#投诉总数
data_brand = data['brand'].value_counts().to_frame().reset_index()
#子品牌个数
data_brand_id = data.groupby('brand')['car_model'].nunique().to_frame().reset_index()#nunique()返回唯一值的个数
```

```python
data_brand1 = data_brand.rename(columns={'index':'brand','brand':'problem_num'}) #重命名
data_brand1.head()#投诉数
```

```python
data_brand_id.head()#子品牌数
```

```python
df2 = pd.merge(data_brand1,data_brand_id, on='brand')
df2.head()
```

```python
df2['brand_mean_problem'] = df2.apply(lambda x:x['problem_num']/x['car_model'],axis=1)
```

```python
df2.set_index('brand').sort_values(by='brand_mean_problem',ascending=False)
```

```python
train_gb = data.groupby('brand')
all_info = {}
for kind, kind_data in train_gb:
    info = {}
    info['brand_mean'] = kind_data['car_model'].value_counts().mean()
    all_info[kind] = info
```

```python
pd.DataFrame(all_info).T.reset_index().sort_values(by='brand_mean',ascending=False)
```
